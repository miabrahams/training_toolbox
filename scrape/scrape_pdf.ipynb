```python
from pypdf import PdfReader
```

```python

reader = PdfReader("/mnt/e/Downloads/1712210088.nothereforever_becomingthebbeg.pdf")

parts = []
last_y = 0


# Use a visitor to ignore header and footer
# https://pypdf.readthedocs.io/en/stable/user/extract-text.html
def visitor_body(text, cm, tm, font_dict, font_size):
    global last_y
    y = tm[5]
    print(text)
    print(tm)
    print(cm)
    if y > 60 and y < 720:
        if (y - last_y < -20):
            # Add a newline between paragraphs
            parts.append("\n")
        last_y = y
        parts.append(text)
    else:
        return
        # print("Deleted: " + text.replace("\n", " "))
        # print(tm)

output = []
for page in reader.pages:
    parts = []
    page.extract_text(visitor_text=visitor_body)
    text_body = "".join(parts)
    output.append(text_body)

print(output)


with open("/mnt/e/Downloads/bbeg.txt", "w") as f:
    f.write("\n\n".join(output))
```
